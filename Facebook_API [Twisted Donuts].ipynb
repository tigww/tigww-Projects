```python
from facebook_business.api import FacebookAdsApi
from facebook_business.adobjects.adaccount import AdAccount
from facebook_business.adobjects.adset import AdSet
from facebook_business.adobjects.adsinsights import AdsInsights
from facebook_business.adobjects.adreportrun import AdReportRun


my_app_id = 'app_id'
my_app_secret = None
my_access_token = 'access_token'
FacebookAdsApi.init(my_app_id, my_app_secret, my_access_token)
account = AdAccount('act_number')

adinsight_fields = [
   AdsInsights.Field.account_id,
   AdsInsights.Field.account_name,
   AdsInsights.Field.campaign_name,
   AdsInsights.Field.campaign_id,
   AdsInsights.Field.impressions,
   AdsInsights.Field.spend,
   AdsInsights.Field.ad_id,
   AdsInsights.Field.ad_name,
   AdsInsights.Field.adset_id,
   AdsInsights.Field.adset_name,
   AdsInsights.Field.adset_start,
   AdsInsights.Field.unique_clicks,
   AdsInsights.Field.clicks,
   AdsInsights.Field.conversions,
   AdsInsights.Field.cpc,
   AdsInsights.Field.cpm,
   AdsInsights.Field.ctr,
   AdsInsights.Field.date_start,
   AdsInsights.Field.date_stop,
   AdsInsights.Field.objective,
   AdsInsights.Field.optimization_goal,
   AdsInsights.Field.outbound_clicks,
]

adsinsights = account.get_insights(
    params={'date_preset':'maximum','level':'ad'},
    fields = adinsight_fields)

print(adsinsights)
```

```python
# Accessing first item in the list which is an adInsights object and we're taking the dictionary form of that object
unwrapped_insight = insights[0].__dict__ #insights = list(adsinsights)
# Converting into a dataframe and transposing the data
unwrapped_metrics = pd.DataFrame(list(unwrapped_insight['_data'])).T

# How to name the first row as a column
columns = unwrapped_metrics.iloc[0]
# Rename column names with the first row of data
column_names = list(columns.values)
# Create empty dataframe with column names
df = pd.DataFrame(columns = column_names)

#Cycle through all insights
for insight in insights:
    #Take the dictionary value and append it into the dataframe
    insight_dict = insight.__dict__['_data']
    df = df.append(insight_dict, ignore_index=True)

df
```

```python
df['account_name'] = df['account_name'].map(lambda x: x.rstrip(',USD'))
```

```python
df['account_name'] = df['account_name'].map(lambda x: x.rstrip(','))
```

```python
df['cpm'] = df['cpm'].astype(float).round(2)
```

```python
df['cpc'] = df['cpc'].astype(float).round(2)
```

```python
df['ctr'] = df['ctr'].astype(float).round(2)
```
